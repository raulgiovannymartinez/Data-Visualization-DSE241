This notebook obtains all of the links required to download all of the CSV files from the RadNet database.

## Load Libraries

```python
from bs4 import BeautifulSoup
from time import sleep
import requests
import pandas as pd
```

## Find all links in RadNet URL
This section obtains the links which contain "csv" in the string, it also excludes certain longer links which do not apply by limiting to 100 characters.

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.epa.gov/radnet/radnet-csv-file-downloads/'
reqs = requests.get(url)
soup = BeautifulSoup(reqs.text, 'html.parser')
 
urls = []
for link in soup.find_all('a'):
    try:
        if 'csv' in link.get('href') and len(link.get('href'))<100:
            urls.append(link.get('href'))
            print(link.get('href'))
    except:
        pass
```

## Number of links obtained

```python
print('Number of Files Downloaded: ', len(urls))
```

## Fix 'space' and replace with "%20"
Some links when downloaded remove a "%20" that's included in the URL and are replaced with a space. Here any space is replaced with %20 to make the link valid.

```python
url_list = []
for link in urls:
    if ' ' in link:
        url_list.append(link.replace(' ','%20'))
    else:
        url_list.append(link)
```

## Save URLs to CSV "all_links.csv"

```python
links_df = pd.DataFrame(url_list,columns=["link"])
links_df
```

```python
links_df.to_csv('all_links.csv',index=False)
```
